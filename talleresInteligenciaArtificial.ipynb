Predecir la velocidad de un automóvil de 9 años, para ello genere las matrices con datos nuevos, grafíquelos y tome dos ejemplos que permitan visualizar la información.

Para cada ejemplo muestre el valor de r, haga que uno de estos ejemplos que usted plantee cumpla con la relacion y el otro No.

```
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
years = np.random.randint(0, 21, 100)
speed_example1 = 180 - 4 * years + np.random.normal(0, 5, 100)
speed_example2 = np.random.randint(100, 200, 100)

# Calcular coeficientes de correlación (valores r)
r_example1 = np.corrcoef(years, speed_example1)[0][1]
r_example2 = np.corrcoef(years, speed_example2)[0][1]

# Graficar datos
plt.scatter(years, speed_example1, label=f"Ejemplo 1 (r = {r_example1:.2f})")
plt.scatter(years, speed_example2, label=f"Ejemplo 2 (r = {r_example2:.2f})")
plt.xlabel("Años de uso")
plt.ylabel("Velocidad (km/h)")
plt.legend()
plt.show()

# Predecir velocidad para un automóvil de 9 años
car_age = 9
predicted_speed_example1 = 180 - 4 * car_age  # Basado en la relación lineal generada para el ejemplo 1
print(f"Velocidad predicha para el ejemplo 1 (automóvil de {car_age} años): {predicted_speed_example1:.2f} km/h")
```

De acuerdo a lo visto para la demostración de regresion polinomial, tome valores donde esta no cumple realizando su gráfica e interpretación con sus respectivos valores

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generar datos aleatorios
np.random.seed(42)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 100, 100)

# Creación del modelo de regresión polinomial (grado 2)
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_features.fit_transform(x[:, np.newaxis])

# Ajuste del modelo
model = LinearRegression()
model.fit(x_poly, y)

# Predecir valores
y_pred = model.predict(x_poly)

# Calcular coeficiente de determinación R^2
r2 = r2_score(y, y_pred)

# Graficar datos y modelo
plt.scatter(x, y, label="Datos")
x_sorted = np.sort(x)
y_sorted = model.predict(poly_features.transform(x_sorted[:, np.newaxis]))
plt.plot(x_sorted, y_sorted, "r", label=f"Modelo (R^2 = {r2:.2f})", linewidth=2)
plt.xlabel("Variable independiente")
plt.ylabel("Variable dependiente")
plt.legend()
plt.show()
```
